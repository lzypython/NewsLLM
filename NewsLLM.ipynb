```python
import json  

def count_entries_in_json_object(file_path):  
    try:  
        with open(file_path, 'r', encoding='utf-8') as f:  
            data = json.load(f)  # 读取JSON数据  
            if isinstance(data, dict):  # 检查数据是否为字典  
                return len(data)  # 返回字典的键的数量  
            else:  
                print("JSON数据不是一个对象。")  
                return 0  
    except Exception as e:  
        print(f"发生错误: {e}")  
        return 0  

# 使用示例  
file_path = 'data.jsonl'  # 替换为你的JSON文件路径  
entry_count = count_entries_in_json_object(file_path)  
print(f"JSON文件中有 {entry_count} 条数据。")
```
